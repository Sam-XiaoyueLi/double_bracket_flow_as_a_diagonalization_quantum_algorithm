```python
from simple_exact_diagonalization_routines.local_matrix_class import *
from scipy.linalg import expm
import matplotlib.pyplot as plt

# System size
L = 3

# Create objects of Pauli operators
X = X_class(L)
Y = Y_class(L)
Z = Z_class(L)
S_plus = S_plus_class(L)
S_minus = S_minus_class(L)

# Trivial initialization of the right size matrix
H_TFIM = 0*np.eye( 2**L ) 

for x in range(L-1):
    H_TFIM = H_TFIM + X.at(x+1).dot(X.at(x+2)) + Z.at(x+1)
H_TFIM = H_TFIM + Z.at(L)    
    
```

```python
from class_double_bracket_flow import *


flow = double_bracket_flow( H_TFIM )

flow.flow_step_max =1
flow.flow_step_min = 0
flow.nmb_search_points_minimizing_s_search = 100
flow.please_use_binary_search  =False

flow.nmb_flow_steps = 15
flow.flow_forwards( states = np.eye(2**L).tolist() )
```

```python
flow2 = double_bracket_flow( H_TFIM )
flow2.please_be_exhaustively_verbose = False
flow2.please_be_verbose = True
ini_state = np.kron( np.kron([0,1],[1,0]), [0,1])
flow2.compute_observables_during_flow( initial_state=ini_state )
flow2.flow_step_max =1
flow2.flow_step_min = 0
flow2.nmb_search_points_minimizing_s_search = 200
flow2.please_use_binary_search  = False
flow2.nmb_flow_steps = 15
flow2.flow_via_best_Z_search( states = np.eye(2**L).tolist(), please_also_check_canonical_bracket=True )
```
